```python
PassingTDs = []
Ints = []
rushattempts = []
rushyards = []
fumbles = []
FantasyPoints = []
PassingYards = []
names = []
Week = []
Position = []
Year = []
Game = []
Name = []
RushTDs = []
```

```python
browser = Browser('chrome')
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)
url1= 'https://www.footballdb.com/fantasy-football/index.html?pos=QB&yr=2023&wk=1&key=b6406b7aea3872d5bb677f064673c57f'
driver.get(url1)
browser.visit(url1)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
div = soup.find('div', class_ = 'table-scroller') 
```

```python
table = div.find('table')
body = table.find('tbody')
TR = body.find_all('tr')
for t in TR:
    TD = t.find('a').text
    names.append(TD)
    Week.append('1')
    Position.append('QB')
    Year.append('2023')
    TD = t.find_all('td')
    for x in range(2,len(TD),18):
        FantasyPoints.append(TD[x].text)
    for x in range(5,len(TD),18):
        PassingYards.append(TD[x].text)
    for x in range(6,len(TD),18):
        PassingTDs.append(TD[x].text)
    for x in range(7,len(TD),18):
        Ints.append(TD[x].text)
    for x in range(9,len(TD),18):
        rushattempts.append(TD[x].text)
    for x in range(10,len(TD),18):
        rushyards.append(TD[x].text)
    for x in range(17,len(TD),18):
        fumbles.append(TD[x].text)
    for x in range(1,len(TD),18):
        Game.append(TD[x].text)
    for x in range(0,len(TD),18):
        Name.append(TD[x].text)
```
