```python
CompleteDF['Fantasy Points'] = CompleteDF['Fantasy Points'].astype(int)
```

```python

CompleteDF['PassingTDs'] = CompleteDF['PassingTDs'].astype(int)
CompleteDF['Receptions'] = CompleteDF['Receptions '].astype(int)
CompleteDF['Receiving Yards'] = CompleteDF['Receiving Yards'].astype(int)
CompleteDF['Receiving TDs'] = CompleteDF['Receiving TDs'].astype(int)
```

```python
CompleteDF
```

```python
for index, row in CompleteDF.iterrows():
    if not np.isnan(row['PassingYards']):
        CompleteDF.at[index, 'PassingYards'] = int(row['PassingYards'])
```

```python
CompleteDF.to_csv('CompleteDataSetWeekasStringANDFPTSasINT.csv')
```
