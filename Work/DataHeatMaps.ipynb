```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a pandas DataFrame
FFpoints = pd.read_csv("Work/excels/CompleteDataSetFinal.csv")
FFpoints
```

```python
# Print the column names to see if they match the ones used in the code
print(FFpoints.columns)
```

```python
 #Selecting relevant columns
selected_columns = ['FantasyPoints', 'PassingYards', 'PassingTDs', 'Ints', 'RushAttemps', 
                   'RushYards', 'Fumbles', 'RushTDs', 'ReceivingYards', 'ReceivingTDs']
```

```python
# Drop any rows with missing values
filtered_df = FFpoints.dropna()

# Filter the DataFrame to include only numeric columns
filtered_df = filtered_df[selected_columns]
filtered_df
```

```python
# Calculate the correlation matrix
correlation_matrix = filtered_df.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap = 'coolwarm', vmin = -1, vmax = 1, center = 0, annot=True, fmt=".2f", square=True, linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Heatmap showcasing one position at a time
# Define the list of positions
positions = ['QB', 'RB', 'WR']

# Define the selected columns
select_columns = ['FantasyPoints', 'PassingYards', 'PassingTDs', 'Ints', 'RushAttemps', 
                    'RushYards', 'Fumbles', 'RushTDs', 'ReceivingYards', 'ReceivingTDs']

# Loop through each position
for position in positions:
    # Filter the DataFrame for the current position
    filtered_df_position = FFpoints[FFpoints['Position'].str.upper() == position]
    
    # Filter the DataFrame to include only numeric columns
    filtered_df_position = filtered_df_position[select_columns]
    
    # Calculate the correlation matrix
    correlation_matrix = filtered_df_position.corr()
    
    # Create the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, cmap='viridis', vmin=-1, vmax=1, center=0, annot=True, fmt=".2f", square=True, linewidths=.5)
    plt.title(f'Correlation Heatmap for Position: {position}')
    plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = filtered_df.corr()

# Create the clustermap
sns.clustermap(correlation_matrix, cmap='viridis', vmin=-1, vmax=1, center=0, annot=True, fmt=".2f", linewidths=.5)

# Display the plot
plt.title('Clustermap of Correlation Matrix')
plt.show()
```
