```python
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float

from sqlalchemy.ext.declarative import declarative_base
# Create a base class for declarating class definitions to produce Table objects
Base = declarative_base()
```

```python
class FantasyFootball(Base): #import table from database
    __tablename__ = 'fantasydata'
    
    id = Column(Integer, primary_key=True)
    Week = Column(Float)
    Name = Column(String)
    Position = Column(String)
    Year = Column(Integer)
    Game = Column(String)
    FantasyPoints = Column(Integer)
    PassingYards = Column(Float)
    PassingTDs = Column(Float)
    Ints = Column(Integer)
    RushAttempts = Column(Integer)
    RushYards = Column(Float)
    Fumbles = Column(Integer)
    RushTDs = Column(Float)
    Receptions = Column(Float)
    ReceivingYards = Column(Float)
    ReceivingTDs = Column(Float)
    
```

```python
engine = create_engine('sqlite:///football.db') #import the right data 
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
from sqlalchemy import func
session = Session(bind=engine)
```

```python
results = session.query(FantasyFootball.Name, FantasyFootball.Year, FantasyFootball.Week, FantasyFootball.FantasyPoints, FantasyFootball.RushYards, FantasyFootball.RushTDs, FantasyFootball.ReceivingYards, FantasyFootball.ReceivingTDs) \
                 .filter(FantasyFootball.Position == 'RB') \
                 .order_by(FantasyFootball.Name) \
                 .all() #create a query to pull indiviual games 
```
